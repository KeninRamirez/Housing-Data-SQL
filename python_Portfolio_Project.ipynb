```python
#Importing Libraries
import pandas as pd
import seaborn as sns
import numpy as np
import os

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')

%matplotlib inline

#Import/Read Data

pwd = os.getcwd()
filepath = pwd + '\\movies.csv'

df = pd.read_csv(filepath)
df
```

```python
#Check any missing data

for i in df.columns:
    pct_missing = np.mean(df[i].isnull())
    print('{} - {}%'.format(i, pct_missing))
```

```python
#Remove missing data

df = df.dropna()
```

```python
#Data Types

df.dtypes
```

```python
#Change data type for float columns into int

df[['budget','gross','votes']] = df[['budget','gross','votes']].astype('int64')
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
#Splitting 'Released' Column

df[['released','country_release']] = df['released'].str.split('\)',n=1,expand = True)
df[['released','country_release']] 
```

```python
df['country_release'] = df['country_release'].str.replace(')','')
df['country_release']
```

```python
df[['Date Released','Year Released']] = df['released'].str.split(', ', expand=True)
df = df.drop(columns='released')
```

```python
#Remove duplicates

df = df.drop_duplicates()

```

```python
#Scatter plot budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
#Correlation gross vs budget

corr_matrix = df.corr(method='pearson')
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
